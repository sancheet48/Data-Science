**Task : Create a decision Tree Classifier and visualize it graphically**

```python
# importing libraries
from IPython import display

import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
from sklearn import tree


import warnings
warnings.filterwarnings('ignore')
```

```python
# Working with the data
df = pd.read_csv('IRIS_Dataset.csv')
df
```

```python
df = df.iloc[: , 1:]
df
```

```python
X = df[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']].values
print(X)
display.Javascript("google.colab.output.setIframeHeight('300px');")
```

```python
y = load_iris().target
y
```

```python
iris = load_iris()
```

```python
# Splitting the data into train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)
```

```python
# fit the classifier
clf = tree.DecisionTreeClassifier(max_depth=3, random_state=10)
clf.fit(X_train, y_train)
```

```python
#Plotting the decision tree classifier
fig = plt.figure(figsize=(20,18))
tree.plot_tree(clf)
plt.show()
```

```python
fig = plt.figure(figsize=(20,18))
tree.plot_tree(clf,
               feature_names = iris.feature_names, 
               class_names=iris.target_names,
               rounded=True, 
               filled = True)
plt.show()
```

```python
# Visualizing using dtreeviz on train dataset
from dtreeviz.trees import dtreeviz
viz = dtreeviz(clf, 
               x_data=X_train,
               y_data=y_train,
               target_name='class',
               feature_names=iris.feature_names, 
               class_names=list(iris.target_names), 
               title="Decision Tree - Iris data set")
viz
```

```python
# Checking the accuracy score on test set
clf.score(X_test, y_test)
```

```python
# Visualizing using dtreeviz on test dataset
viz = dtreeviz(clf, 
               x_data=X_test,
               y_data=y_test,
               target_name='class',
               feature_names=iris.feature_names, 
               class_names=list(iris.target_names),
            
               title="Decision Tree - Iris data set")
viz
```
